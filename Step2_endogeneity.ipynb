```python

selected_boroughs = [
    'Camden', 'Islington', 'Hackney', 'Tower Hamlets', 'Greenwich',
    'Southwark', 'Lambeth', 'Hammersmith and Fulham', 'Kensington and Chelsea',
    'Westminster', 'Brent'
]

# Filter the rent DataFrame to include only the selected boroughs
filtered_rent_df = df[df['Borough'].isin(selected_boroughs)]
```

```python
crime_df = pd.read_csv(file_url)

# Filter for 'SEXUAL OFFENCES'
sexual_offences_df = crime_df[crime_df['MajorText'] == 'SEXUAL OFFENCES']

# Group by 'BoroughName' and 'Year' and sum 'Value'
crime_summary = sexual_offences_df.groupby(['BoroughName', 'Year', 'CrimeRate']).sum().reset_index()

# Rename columns
crime_summary = crime_summary.rename(columns={
    'BoroughName': 'borough',
    'Year': 'year',
    'Value': 'sex_crimes'
})

# Assign the processed DataFrame back to crime_df
crime_df = crime_summary.copy()

print("Crime data loaded, filtered, grouped, and columns renamed. Displaying first 5 rows of crime_df:")
print(crime_df.head())
```

```python
rent_df = pd.read_csv(file_url2)

# Rename columns
rent_df = rent_df.rename(columns={
    'Borough': 'borough',
    'Mean_Monthly_Rent': 'rent_price'
})

print("Rent data loaded and columns renamed. Displaying first 5 rows of rent_df:")
print(rent_df.head())
```

```python
merged_crime_rent_df = pd.merge(crime_df, rent_df, on='borough', how='inner')

print("Merged DataFrame (merged_crime_rent_df) head:")
print(merged_crime_rent_df.head())

print("\nMerged DataFrame (merged_crime_rent_df) info:")
merged_crime_rent_df.info()
```

**Test for endogeneity, correlation analysis**

```python
df_all_ages = df_all_ages.rename(columns={'Borough': 'borough'})
print("Column 'Borough' renamed to 'borough' in df_all_ages.")
print("Updated columns:", df_all_ages.columns.tolist())
```

```python
final_merged_df = pd.merge(merged_crime_rent_df, df_all_ages[['borough', 'All ages']], on='borough', how='inner')
print("Merged DataFrame (final_merged_df) head:")
print(final_merged_df.head())
print("\nMerged DataFrame (final_merged_df) info:")
final_merged_df.info()
```

```python
import statsmodels.api as sm
from scipy.stats import pearsonr

def partial_correlation(df, x, y, controls):
    """Calculate partial correlation between x and y controlling for controls"""
    # Residuals after controlling for other variables
    X_control = sm.add_constant(df[controls])

    # Regress x on controls
    model_x = sm.OLS(df[x], X_control).fit()
    resid_x = model_x.resid

    # Regress y on controls
    model_y = sm.OLS(df[y], X_control).fit()
    resid_y = model_y.resid

    # Correlation of residuals
    return pearsonr(resid_x, resid_y)[0]

# Calculate partial correlation
controls = ['All ages']
partial_corr_crime_rent = partial_correlation(final_merged_df, 'CrimeRate', 'rent_price', controls)

print(f"Partial correlation between CrimeRate and rent_price, controlling for 'All ages': {partial_corr_crime_rent:.3f}")
```
